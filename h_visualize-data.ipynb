##### 그래프의 종류
  
**막대 그래프** : 데이터를 카테고리별로 값을 막대로 표현 할때 사용하는 그래프 형식 (수직 막대 그래프, 수평 막대 그래프)  
> 카테고리별 데이터를 비교하고자 할때 사용 (제품 판매량, 지역별 인구수)  
  
**선 그래프** : 시간의 흐름에 따른 데이터 포인터를 선으로 연결하여 사용하는 그래프 형식  
> 하나의 요소에 대해 시간의 흐름에 따라 변경값을 나타내고자 할때 사용 (기온, 주식)  
  
**파이 차트** : 전체 데이터에 대한 각 요소의 비율을 원형으로 나타낼때 사용하는 그래프 형식  
> 전체에 대한 요소의 비율을 표현하고자 할때 사용 (투표 결과)  
  
**산점도** : 두 변수 간의 상관관계를 표현할때 사용하는 그래프 형식  
> 각 요소간 관계성을 파악할때 사용 (키와 몸무게 상관관계 비교)  
  
**히스토그램** : 범위별로 그룹화하여 각 그룹에 속하는 데이터의 수를 막대로 표현하는 그래프 형식  
> 데이터의 분포를 비교할때 사용 (시험 점수 분포)  
  
**영역 그래프** : 선 그래프의 아래쪽 빈 공간을 채운 그래프  
> 시간의 흐름에 따라 데이터를 비교할 때 사용, 변화량을 강조하고자 할때 사용  
  
**박스 그래프** : 데이터의 분포를 나타낼때 사용하는 그래프 형식  
> 최대값, 최소값, 사분위수로 데이터의 이상치를 판단할때 사용

````python
data = [1, 2, 4, 8, 16, 32]
````

````python
# 그래프이 X축 변경 
plt.title('X 축 추가')
# plot(X축의 리스트, Y축의 리스트)
plt.plot([2020, 2021, 2022, 2023, 2024, 2025], data)
````

````python
x = [2020, 2021, 2022, 2023, 2024, 2025]
````

````python
# 차트 스타일 지정
# plot(X축리스트, Y축리스트, 스타일문자열)
# 지정된 문자열 포멧으로 전달하여야 스타일 지정가능
plt.title('c2-- 스타일 적용')
plt.plot(x, data, 'c2--')
plt.show()
````

````python
# 스타일 상세 조정
# plot() 메서드에 각각에 대한 스타일 매게변수 전달
# marker : 마커 종류 / markersize (ms) : 마커의 크기
# markeredgecolor (mec) : 마커 테두리 색상 
# markeredgewidth (mew) : 마커 선 굵기
# markerfacecolor (mfc) : 마커 내부 색상
# color (c) : 선 색상
# linewidth (lw) : 선 굵기
# linestyle (ls) : 선 스타일
plt.title('상세 스타일 적용')
plt.plot(x, data, c='#00998850', lw=5, ls=':', marker='*', ms=10, mfc='#009988')
plt.show()
````

````python
# 차트 범위 지정
# xlim(), ylim() 메서드로 x축과 y축의 상,하한을 지정할 수 있음
plt.title('차트 범위 지정')
plt.plot(x, data, c='#00998850', lw=5, ls=':', marker='*', ms=10, mfc='#009988')
plt.xlim(2019, 2026)
plt.ylim(-5, 40)
plt.show()
````

````python
# 틱 지정
# xticks(), yticks() 메서드로 틱의 위치나 라벨을 지정 가능
plt.title('틱 지정')
plt.plot(x, data, c='#00998850', lw=5, ls=':', marker='*', ms=10, mfc='#009988')
plt.xticks([2020, 2022, 2024])
plt.yticks([1, 8, 32], ['저점', '중점', '고점'])
plt.show()
````

````python
# 그리드
# grid() 메서드를 이용해서 그리드 표시 여부를 지정 할 수 있음
plt.title('그리드')
plt.plot(x, data, c='#00998850', lw=5, ls=':', marker='*', ms=10, mfc='#009988')
plt.grid(True)
plt.show()
````

````python
# 그래프 겹쳐 그리기
# plot() 메서드를 중첩해서 사용하면 데이터를 겹쳐서 표현 가능
plt.title('차트 겹치기')
plt.plot(x, data, c='#00998850', lw=5, ls=':', marker='*', ms=10, mfc='#009988')
plt.plot(x, list(reversed(data)), c='#88990050', lw=5, ls=':', marker='D', ms=10, mfc='#889900')
plt.show()
````

````python
# 범례 표시 
# legend() 메서드로 범례를 표시할 수 있음 
# loc 매게변수에 위치에 해당하는 정수를 전달하여 위치 지정 가능 
plt.plot(x, data, c='#00998850', lw=5, ls=':', marker='*', ms=10, mfc='#009988', label='A')
plt.plot(x, list(reversed(data)), c='#88990050', lw=5, ls=':', marker='*', ms=10, mfc='#889900', label='B')
plt.title('범례 표시')

plt.legend(loc=7) 
plt.show()
````

````python
# 틱 라벨 지정
# xlabel(), ylabel() 메서드로 x축, y축의 라벨을 표시할 수 있음
plt.title('틱 라벨')
plt.plot(x, data, c='#00998850', lw=5, ls=':', marker='*', ms=10, mfc='#009988', label='A')
plt.plot(x, list(reversed(data)), c='#88990050', lw=5, ls=':', marker='D', ms=10, mfc='#889900', label='B')
plt.xlabel('연도')
plt.ylabel('가치')
plt.show()
````

### 막대 그래프
x축의 값이 분류가 가능한 카테고리 형태일때 bar(), barh() 메서드로 막대 그래프를 표현할 수 있음

````python
plt.title('막대그래프')
plt.bar(['전자기기', '가전제품', '사무용품'], [40, 20, 33], color='#ff000050')
plt.show()
````

````python
plt.title('막대그래프')
# 오차 표현
# xerr, yerr 매개변수로 오차를 표현할 수 있음
plt.barh(['전자기기', '가전제품', '사무용품'], [40, 20, 33], xerr=[4, 2, 3], color='#ff000050')
plt.show()
````

### 스템 플롯
기준점을 기준으로 오차범위를 표현하고자 할때 사용하는 차트  

````python
import numpy as np
x = np.linspace(0.1, 2 * np.pi, 10)
plt.title('스템 플룻')
plt.stem(x, np.cos(x))
plt.show()
````

### 파이차트
각 분류가 전체에서 차지하는 비중을 표현하고자 할때 사용하는 차트

````python
plt.title('파이 차트')
plt.pie([25, 40, 35], labels=['전자기기', '가전제품', '사무용품'], autopct='%1.1f%%', colors=['#ff000050', '#00ff0050', '#0000ff50'], explode=[0, 0.1, 0])
plt.show()
````

### 컨투어 플롯
3차원 자료에서 x,y 입력과 z출력이 존재할 때, z를 선과 색상으로 표현하는 그래프

````python
x = np.linspace(-3, 3, 256)
y = np.linspace(-3, 3, 256)
xx, yy = np.meshgrid(x, y)
zz = (1 - xx / 2 + xx ** 5 + yy ** 3) * np.exp(-xx ** 2 - yy ** 2)

plt.title('등고선 그래프')
plt.contour(xx, yy, zz, colors='black')
plt.contour(xx,yy,zz)
plt.show()
````
